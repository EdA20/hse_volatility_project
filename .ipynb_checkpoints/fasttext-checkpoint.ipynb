```python
import os

import numpy as np
import pandas as pd
```

```python
import fasttext
import fasttext.util
```

```python
help(fasttext.util.download_model)
```

```python
# fasttext.util.download_model('ru', if_exists='ignore')
ft = fasttext.load_model('cc.ru.300.bin')
```

```python
ft['привет']
```

```python
def get_tweet_embedding(lemmas, model, embedding_size=200):
    
    res = np.zeros(embedding_size)
    cnt = 0
    for word in lemmas.split():
        if word in model:
            res += np.array(model[word])
            cnt += 1
    if cnt:
        res = res / cnt
    return res
```

```python
x = 'привет всем слушателям курса'
get_tweet_embedding(x, model=ft, embedding_size=300)
```

```python
BACKTEST_CCY = 'USDRUB'
BACKTEST_DAYS = 5

PATH_TEXTS = 'data/telegram'
PATH_OPT_PNL = 'data/pnl'
```

```python
# Get all available data_sources
sources = os.listdir(PATH_TEXTS)
sources
```

```python
# Generate dataframes
data_sources = []

for s in sources:
    if s != '.gitignore':
        source_data = pd.read_csv(f'{PATH_TEXTS}/{s}')
        data_sources.append((s.split('.')[0], source_data))
```

```python
# Create target variable dataframe
pnl = pd.read_csv(f'{PATH_OPT_PNL}/Backtest_{BACKTEST_CCY}_{BACKTEST_DAYS}_days.txt')
pnl['date_start'] = pd.to_datetime(pnl['date_start']).dt.strftime('%Y-%m-%d')
pnl['pnl_sign'] = pnl['pnl'].apply(lambda x: 1 if x >= 0 else 0)
pnl
```

```python
# Get balance of the sample
pnl['pnl_sign'].sum() / pnl.shape[0]
```

```python
data_sources[0][1]
```

```python
df = pnl
for source_name, source_dataframe in data_sources:
    df[f'{source_name}'] = ''
    
    i = 0
    texts = ''
    for _, row in source_dataframe.iterrows():
        if i >= df.shape[0]:
            break

        if pd.to_datetime(row['date']).strftime('%Y-%m-%d') <= df['date_start'][i]:
            texts += ' ' + row['text']
        else:
            df.at[i, source_name] = texts
            texts = ''
            i += 1
            continue
```

```python
df = df.iloc[1:]
df
```

```python
df = df[~df.isin([''])].dropna(axis=1, how='all')
```

```python
df_final = df.drop(['cbonds', 'signal', 'rshb_invest', 'Alfa_Wealth'], axis=1)
# df_final = df
df_final
```

```python
X = df_final.drop(['date_start', 'pnl', 'pnl_sign'], axis=1)
y = df_final['pnl_sign']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2, random_state=121)
```

```python
X_train.shape
```

```python
X_train
```

```python
X_train['mmi'] = X_train['mmi'].apply(lambda x: get_tweet_embedding(x, model=ft, embedding_size=300))
print('train done')

X_val['mmi'] = X_val['mmi'].apply(lambda x: get_tweet_embedding(x, model=ft, embedding_size=300))
```

```python
import sklearn
```

```python
print(list(X_train['mmi'].values))
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

clf = LogisticRegression(random_state=121)
clf.fit(X_train['mmi'], y_train)

pred = clf.predict(X_val['mmi'])
accuracy_score(pred, y_val)
```
