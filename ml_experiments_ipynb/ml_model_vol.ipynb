```python
df = pnl
for source_name, source_dataframe in data_sources:
    df[f'{source_name}'] = ''
    
    i = 0
    texts = ''
    for _, row in source_dataframe.iterrows():
        if i >= df.shape[0]:
            break

        if pd.to_datetime(row['date']).strftime('%Y-%m-%d') <= df['date_start'][i]:
            texts += ' ' + row['text']
        else:
            df.at[i, source_name] = texts
            texts = ''
            i += 1
            continue
```

```python
df = df.iloc[1:]
df
```

```python
df = df[~df.isin([''])].dropna(axis=1, how='all')
```

```python
class Word2VecTransformer:
    
    def __init__(self, w2v_model, word_pattern):
        
        self.w2v_model = w2v_model
        self.word_pattern = word_pattern
        self.re = re.compile(pattern=self.word_pattern)
        
    def fit(self, X):
        return self
    
    def transform(self, X):
        
        X_transformed = np.zeros((len(X), self.w2v_model.wv.vector_size))
        for i, title in enumerate(X):
            
            title_vector = np.zeros((self.w2v_model.wv.vector_size,))
            tokens = self.re.findall(title.lower())
            for token in tokens:
                if token in self.w2v_model.wv.key_to_index:
                    title_vector += self.w2v_model.wv.get_vector(token)
                    
            X_transformed[i] = title_vector
        
        return X_transformed
```

```python
import re
from gensim.models.word2vec import Word2Vec
from sklearn.linear_model import SGDClassifier

# From CountVectorizer() (default pattern)
WORD_PATTERN = '(?u)\\b\\w\\w+\\b'

reg_exp = re.compile(pattern=WORD_PATTERN)

X_train_transformed = np.empty(shape=(len(y_train), 100))
X_val_transformed = np.empty(shape=(len(y_val), 100))
for i in range(len(X_train.columns)):
    sentences = [reg_exp.findall(s.lower()) for s in X_train.iloc[:,i].astype(str)]
    w2v_model = Word2Vec(sg=1,)
    w2v_model.build_vocab(sentences)
    w2v_model.train(
        sentences,
        total_examples=w2v_model.corpus_count,
        epochs=20,
        compute_loss=True
    )
    
    w2v_transformer = Word2VecTransformer(w2v_model=w2v_model, word_pattern=WORD_PATTERN)

    train_w2v = w2v_transformer.transform(X_train.iloc[:,i].astype(str))
    val_w2v = w2v_transformer.transform(X_val.iloc[:,i].astype(str))
    
    print(X_train_transformed.shape)
    
    if i == 0:
        X_train_transformed = train_w2v
        X_val_transformed = val_w2v
    else:
        X_train_transformed = np.append(X_train_transformed, train_w2v, axis=1)
        X_val_transformed = np.append(X_val_transformed, val_w2v, axis=1)

X_train_transformed.shape
```
